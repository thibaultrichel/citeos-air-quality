```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
classifiers = {
    "Random Forest": {
        "accuracy": 0.81,
        "f1_score": 0.73
    },
    "XGBoost": {
        "accuracy": 0.76,
        "f1_score": 0.69
    },
    "LightGBM": {
        "accuracy": 0.75,
        "f1_score": 0.64
    },
    "SVC": {
        "accuracy": 0.71,
        "f1_score": 0.59
    }
}

regressors = {
    "LSTM_with_target": {
        "MAE": 0.54
    },
    "LSTM_no_target": {
        "MAE": 1.2
    },
    "Random Forest": {
        "MAE": 0.68,
    }
}
```

```python
c_accs = [classifiers[cl]["accuracy"] for cl in classifiers]
c_f1s = [classifiers[cl]["f1_score"] for cl in classifiers]
r_maes = [regressors[cl]["MAE"] for cl in regressors]
```

```python
f, ax = plt.subplots(2, figsize=(20, 10))

n1 = 4
ind1 = np.arange(n1)
width = 0.2
ax[0].bar(ind1, c_accs, width, color="b", label="Accuracy")
ax[0].bar(ind1 + width, c_f1s, width, color="r", label="F1 Score")
ax[0].set_title("Classifiers performances")
ax[0].set_xticklabels(list(classifiers.keys()))
ax[0].set_xticks(ind1 + width/2)
ax[0].set_xlabel("Classifiers")
ax[0].set_ylabel("Metrics")
ax[0].legend()

ax[1].bar(list(regressors.keys()), r_maes, width, color="g", label="MAE")
ax[1].set_xlabel("Classifiers")
ax[1].set_ylabel("Metrics")
ax[1].legend()
```
