```python
def parse_date(string):
    res = string.replace("T", " ").split(" ")
    date = "-".join([res[0][:4], res[0][4:6], res[0][6:]])
    hour = ":".join([res[1][:2], res[1][2:]])
    return " ".join([date, hour])
```

```python
df = pd.read_csv("./data/data-test-bale-meteoblue-historyplus.csv", skiprows=10, names=["timestamp", "temperature", "humidity", "wind-speed", "wind-direction"])
df.timestamp = df.timestamp.apply(parse_date)
df["day"] = df.timestamp.apply(lambda x: x.split(" ")[0])
df = df[["timestamp", "day", "temperature", "humidity", "wind-speed", "wind-direction"]]
df
```
